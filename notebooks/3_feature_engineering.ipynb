```python
# Impor pustaka
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import os

# Buka data mentah
df = pd.read_csv("../data/raw/social_media_engagement_data.csv")
print("✅ Data dimuat:", df.shape)

# 1. Bersihkan data: hapus duplikat & nilai kosong
df = df.drop_duplicates()
df = df.dropna(subset=['Post_Type', 'Time_Posted', 'Engagement_Rate'])
print("✅ Data dibersihkan:", df.shape)

# 2. Buat fitur baru: Jam, Hari, Bulan
df['Date'] = pd.to_datetime(df['Date'])
df['Hour'] = df['Time_Posted'].str.split(':').str[0].astype(int)
df['Day_of_Week'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.month_name()

# 3. Ubah teks jadi angka (kode)
le = LabelEncoder()
df['Post_Type_Code'] = le.fit_transform(df['Post_Type'])
df['Day_Code'] = le.fit_transform(df['Day_of_Week'])

# 4. Skala angka agar setara
scaler = MinMaxScaler()
df[['Engagement_Rate', 'Hour']] = scaler.fit_transform(df[['Engagement_Rate', 'Hour']])

# 5. Simpan data yang sudah jadi
os.makedirs("../data/processed", exist_ok=True)
df.to_csv("../data/processed/data_siap_latih.csv", index=False)
print("\n💾 Data sudah disimpan di: data/processed/data_siap_latih.csv")
print("\nContoh data akhir:")
print(df[['Post_Type','Hour','Day_of_Week','Engagement_Rate']].head())
```
