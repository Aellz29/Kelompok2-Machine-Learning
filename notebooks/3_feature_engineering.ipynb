```python
# Impor pustaka
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import os

# 1. Buka data mentah
df = pd.read_csv("../data/social_media_engagement_dataset.csv")
print("✅ Data dimuat:", df.shape)

# 2. Bersihkan data: hapus duplikat & nilai kosong
df = df.drop_duplicates()
df = df.dropna(subset=['Content_Type', 'Hour_of_Day', 'Engagement_Rate'])
print("✅ Data dibersihkan:", df.shape)

# 3. Skala angka agar setara (Tugas Feature Engineering - Kolom Tunggal Tanpa 's')
scaler = MinMaxScaler()
df[['Follower_Count', 'Content_Length', 'Hour_of_Day']] = scaler.fit_transform(
    df[['Follower_Count', 'Content_Length', 'Hour_of_Day']]
)
print("✅ Feature Scaling berhasil menggunakan MinMaxScaler!")

# 4. Ubah teks kategori jadi angka (Label Encoding)
le = LabelEncoder()
df['Platform_Code'] = le.fit_transform(df['Platform'])
df['Content_Type_Code'] = le.fit_transform(df['Content_Type'])

# 5. Simpan data yang sudah jadi ke folder data (SUDAH DIGANTI BIAR AMAN)
os.makedirs("../data", exist_ok=True)
df.to_csv("../data/data_scaled_widya.csv", index=False)
print("\n💾 Data sudah disimpan di: data/data_scaled_widya.csv")
```
