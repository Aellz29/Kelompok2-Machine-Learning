```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_sosmed_bersih.csv")

df.head()
```

```python
df["Total_Engagement"] = (
    df["Likes"] +
    df["Comments"] +
    df["Shares"] +
    df["Saves"]
)

df[["Content_Type","Total_Engagement"]].head()
```

```python
df["Content_Type"].value_counts()
```

```python
content_engagement = (
    df.groupby("Content_Type")["Total_Engagement"]
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=content_engagement.index,
    y=content_engagement.values
)

plt.title("Rata-rata Engagement Berdasarkan Jenis Konten")
plt.xlabel("Jenis Konten")
plt.ylabel("Rata-rata Engagement")

plt.xticks(rotation=45)

plt.show()
```

Analisis Hubungan Jenis Konten terhadap Engagement

Berdasarkan hasil analisis rata-rata Total Engagement berdasarkan jenis konten, ditemukan bahwa konten Duet memiliki tingkat engagement tertinggi dengan nilai rata-rata sebesar 31.390, diikuti oleh Stitch sebesar 30.508 dan Video sebesar 17.081. Hasil ini menunjukkan bahwa konten yang bersifat interaktif dan melibatkan partisipasi pengguna cenderung menghasilkan tingkat engagement yang lebih tinggi.

Sebaliknya, konten Document dan Article memiliki tingkat engagement terendah dengan rata-rata masing-masing sebesar 1.479 dan 1.577. Hal ini menunjukkan bahwa pengguna media sosial lebih tertarik pada konten visual dan interaktif dibandingkan konten berbasis dokumen atau artikel yang memerlukan waktu lebih lama untuk dikonsumsi.

Temuan ini mengindikasikan bahwa pemilihan jenis konten merupakan faktor penting dalam meningkatkan engagement pengguna pada platform media sosial.

```python
likes_content = (
    df.groupby("Content_Type")["Likes"]
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=likes_content.index,
    y=likes_content.values
)

plt.title("Rata-rata Likes Berdasarkan Jenis Konten")
plt.xlabel("Jenis Konten")
plt.ylabel("Likes")

plt.xticks(rotation=45)

plt.show()
```

### Analisis Jenis Konten terhadap Jumlah Likes

Berdasarkan hasil analisis, konten Duet memperoleh rata-rata jumlah likes tertinggi yaitu sebesar 26.245, diikuti oleh Stitch sebesar 25.421 dan Video sebesar 13.913. Hasil ini menunjukkan bahwa pengguna media sosial cenderung memberikan apresiasi lebih tinggi terhadap konten yang bersifat interaktif dan kreatif.

Sementara itu, konten Document dan Article memiliki rata-rata likes paling rendah. Hal ini mengindikasikan bahwa konten berbasis teks atau dokumen kurang menarik perhatian pengguna dibandingkan konten visual maupun audiovisual.

Temuan ini menunjukkan bahwa format penyajian konten memiliki pengaruh yang signifikan terhadap tingkat ketertarikan pengguna yang diukur melalui jumlah likes.

```python
engagement_rate_content = (
    df.groupby("Content_Type")["Engagement_Rate"]
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=engagement_rate_content.index,
    y=engagement_rate_content.values
)

plt.title("Rata-rata Engagement Rate Berdasarkan Jenis Konten")
plt.xlabel("Jenis Konten")
plt.ylabel("Engagement Rate")

plt.xticks(rotation=45)

plt.show()
```

Hasil analisis menunjukkan bahwa Stitch memiliki rata-rata Engagement Rate tertinggi sebesar 45,38%, diikuti oleh Duet sebesar 41,63% dan Video sebesar 20,41%. Tingginya nilai engagement rate pada jenis konten tersebut menunjukkan bahwa pengguna lebih aktif berinteraksi dengan konten yang bersifat kolaboratif, interaktif, dan menarik secara visual.

Sebaliknya, konten Document dan Article memiliki engagement rate paling rendah, yaitu masing-masing sebesar 1,18% dan 1,45%. Hal ini mengindikasikan bahwa konten berbasis teks atau dokumen cenderung menghasilkan interaksi yang lebih sedikit dibandingkan konten audiovisual.

Secara umum, hasil ini menunjukkan bahwa penggunaan format konten yang interaktif dapat menjadi strategi yang efektif dalam meningkatkan tingkat keterlibatan pengguna di media sosial.
